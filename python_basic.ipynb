```python
#OOP
class car:
  def __init__(self,brand,model,year):
    self.brand=brand
    self.model=model
    self.year=year
  def car_info(self):
    print(self.brand,self.model,self.year)
car1=car("Toyota","Corolla",2025)
car1.car_info()

```

```python
#OOP 2
#2nd oop
class vehicals:
  def __init__ (self,brand,model,color):
    self.brand = brand
    self.model = model
    self.color = color
  def vechicle_info(self):
    print(self.brand,self.model,self.color)
carz = vehicals("lx","2016","gray")
carz.vechicle_info()
```
